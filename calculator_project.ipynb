```python
def addition(num1,num2):
     print(f"{num1} + {num2} = {num1 + num2}")
 
def subtraction(num1,num2):
    print(f"{num1} - {num2} = {num1 - num2}")
    
def multiplication(num1,num2):
    print(f"{num1} * {num2} = {num1 * num2}")
    
def division(num1,num2):
    if num2 == 0:
        print("A number cannot be divided by zero.")
    else:    
        print(f"{num1} / {num2} = {num1 / num2}")
    
def reminder(num1,num2):
    print(f"{num1} % {num2} = {num1 % num2}")
    
def exponent(num1,num2):
    print(f"{num1} ** {num2} = {num1 ** num2}")



while True:
    
    number1 = int(input("enter first number: "))
    print()
    operator = input("enter one operator in below option..\
    \n1.+\n2.-\n3.*\n4./\n5.%\n6.**.\n")
    print()
    number2 = int(input("enter second number: "))
    print()


    
    if (operator == "+"):
        addition(number1,number2)
        print()
    
    elif (operator == "-"):
        subtraction(number1,number2)
        print()
    
    elif (operator == "*"):
        multiplication(number1,number2)
        print()
    
    elif (operator == "/"):
        division(number1,number2)
        print()
    
    elif (operator == "%"):
        reminder(number1,number2)
        print()
    
    elif (operator == "**"):
        exponent(number1,number2)
        print()
    
    else:
        print(f"{operator} is Invalid operator")
        print()


    
    while True:
        
        choice = input("enter one option in below..\n1.'continue'. for continue calculation.\n2.'exit'. for stop calculation.\n").lower()

        
        if choice == "continue".lower():
            break
        
        elif choice == "exit".lower():
            break
        
        else:
            print(f"you enter '{choice}' which is wrong option --> try again and select correct option.")
            print()

    
    if choice == "exit":
        break
    
    else:
        print("continue calculation..")
        print()
```
