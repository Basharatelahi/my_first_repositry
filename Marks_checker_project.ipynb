```python
def Marks_cheaker():
    student_name = input("enter your name: ")

    print(f"Welcome To marks Checker \'{student_name}\',")
    print()

    total_subject = int(input("enter total subject : "))
    print(f"Total subject = {total_subject},")
    print()

    info_store = {}
    for i in range(total_subject):
        paper_name = input("enter subject name: ").lower()
        paper_marks = int(input("enter subject marks: "))
        info_store[paper_name] = paper_marks

    print("Paper marks:")
    for sub_name,sub_marks in info_store.items():
        print(f"{sub_name} = {sub_marks},")
    print()

    marks_List = []
    for marks in info_store.values():
        marks_List.append(marks)
        
    Obtained_marks = sum(marks_List)

    Total_marks = total_subject * 100
    print(f"obtaining marks = ({Obtained_marks}/{Total_marks}),")
    print()

    percentage = (Obtained_marks * 100) / (Total_marks)
    print(f"percentage = {percentage:.2f}%,") 
    print()

    high_marks_sub = ""
    high_marks = 0
    for sub,marks in info_store.items():
        if marks > high_marks:
            high_marks_sub = sub
            high_marks = marks
        print(f"High_marks_subject: {high_marks_sub} = {high_marks},")
        print()

    low_marks_sub = ""
    low_marks = 100
    for sub,marks in info_store.items():
        if marks < low_marks:
            low_marks_sub = sub
            low_marks = marks
            print(f"Low_marks_subject: {low_marks_sub} = {low_marks},")   
            print()

    print(f"Conclusion and Insight is: \'{student_name}\' are strong in ({high_marks_sub}) subject, and weak in ({low_marks_sub}) subject.")

Marks_cheaker()    
```
